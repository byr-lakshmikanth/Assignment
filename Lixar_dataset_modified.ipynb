```python
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('/home/itv000367/kanth/retailer-dataset.csv')
```

```python
df.columns
```

```python
df.rename(columns={'timestamp':'timestamp_started'},inplace=True)
```

```python
df.head()
```

```python
# changing dtype of timestamp_started column as datetime 
df['timestamp_started'] = pd.to_datetime(df['timestamp_started'])
```

```python
df['timestamp_started']
```

```python
# Creating an array to generate random values for minutes
min_arr = np.random.randint(1,10, size=20723)
```

```python
len(min_arr)
```

```python
min_arr
```

```python
# convert an arr in to series
min_series = pd.Series(min_arr)
```

```python
# add a minute column to dataframe df
df['minutes'] = min_series
```

```python
df.head()
```

```python
# add minutes to timestamp_started to obtain timestamp_ended for each event
df['time_added'] = pd.to_timedelta(df['minutes'],'m')
df['timestamp_ended'] = df['timestamp_started'] + df['time_added']
```

```python
df.head()
```

```python
df.drop(['minutes','time_added'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.sort_values(by=['user_id','merchant_id','flyer_id'],inplace=True)
```

```python
df
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df 
```

```python
# write modified dataframe to a csv file
df.to_csv('/home/itv000367/kanth/modified-retailer-dataset.csv')
```
